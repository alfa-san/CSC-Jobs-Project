```python
import sqlite3
import pandas as pd
from pathlib import Path

# Path to your merged CSV
base = Path("C:/Users/HP Omen 15/Documents/[BS Economics]/RESUME/CSC_job_listings")
csv_path = base / "MergedData.csv"

# Load CSV into pandas
df = pd.read_csv(csv_path)

# Connect to SQLite (creates a local DB file)
conn = sqlite3.connect("csc_job_listings.db")

# Write DataFrame to SQL table
df.to_sql("csc_job_listings", conn, if_exists="replace", index=False)
```

```python
# Example SQL query
query = "SELECT * FROM csc_job_listings;"
results = pd.read_sql_query(query, conn)

results
```

```python
query = "PRAGMA table_info(csc_job_listings);"
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT DISTINCT "Position Title"
FROM csc_job_listings
ORDER BY "Position Title"
"""
# Run query and store results in DataFrame
df = pd.read_sql_query(query, conn)
df
```

```python
# Export DataFrame to CSV
output_path = "unique_position_titles.csv"
df.to_csv(output_path, index=False)

print(f"Results saved to {output_path}")
```

```python
query = """
SELECT DISTINCT Agency
FROM csc_job_listings
ORDER BY Agency
"""
# Run query and store results in DataFrame
df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT DISTINCT "Position Title"
FROM csc_job_listings
WHERE Region IN ('Region VII', 'Region VIII')
ORDER BY "Position Title";
"""
# Run query and store results in DataFrame
df = pd.read_sql_query(query, conn)
df
```

```python
# Export DataFrame to CSV
output_path = "UniquePositionsR7R8.csv"
df.to_csv(output_path, index=False)

print(f"Results saved to {output_path}")
```
